## Финансовые показатели 

Далее рассчитаем основные финансовые показатели, которые будут использованы в исследовании:

Показатели рентабельности:

1. ROE – рентабельность капитала рассчитывается как отношение чистой прибыли к собственным средствам банка. ROE иногда может быть высоким из-за сильной долговой нагрузки (левериджа). Для банков ROE > 15% часто сопровождается повышенными рисками.

2. ROA – рентабельность активов рассчитывается как отношение чистой прибыли к общим активам банка. Чем выше ROA, тем больше активы генерируют прибыль. Для банков низкий ROA может быть связан с большими резервами под кредитные риски или низкой маржинальностью операций.

3. NIM – чистая процентная маржа рассчитывается как отношение чистого процентного дохода (разница между процентными доходами и процентными расходами) к кредитному портфелю банка. Отражает отдачу на вложенные активы в основной деятельности банка, то есть в чистом процентном доходе.


Показатели бизнес-стратегии банка (структуры баланса):

4. LTD – отношение выданных кредитов к привлеченным средствам клиентов. Показывает, насколько банк зависит от привлеченных средств (депозитов) для финансирования кредитов. Чем ближе значение к 100%, тем больше средств клиентов тратится на основную деятельность банка. Если LTD < 70% - банк мало кредитует, LTD в диапазоне 70-90% - оптимальное значение, LTD – 90-100% говорит о повышенном риске, т.к. банк привлекает короткие деньги, а если LTD > 100% – это признак риска дефицита ликвидности.

5. Доля кредитного портфеля в активах банка (loan_share) рассчитывается как отношения выданных кредитов к общим активам банка. 

6. Также рассчитывается отношение средств клиентов (депозитов) к общим активам банка (dep_to_assets).


Показатели кредитного риска:

7. NPLR – доля просроченных кредитов в кредитном портфеле банка. Чем выше NPLR, тем выше кредитный риск. Если NPLR < 3% - низкий кредитный риск, NPLR в диапазоне 5%-10% говорит о высоком кредитном риске, а NPLR > 10% – критический уровень (угроза капиталу банка).


Меры общего риска:

8. CAR – отношение собственных средств к активам. Показывает долю активов, финансируемых за счет собственного капитала, а не заемных средств. Чем выше показатель, тем ниже зависимость от заемного финансирования. CAR равный 10% и выше – это запас прочности, CAR < 5% – тревожный сигнал (высокий левередж).

```python
data['ROE'] = (data['Net_Profit'] / data['EQUITY']) *100
data['ROA'] = (data['Net_Profit'] / data['ASSETS']) *100
data['NIM'] = (data['Interest_Income'] - data['Interest_Expense'])/data['LOAN_PORTFOLIO']

data['LTD'] = data['LOAN_PORTFOLIO'] / data['DEPOSITS']
data['loan_share'] = data['LOAN_PORTFOLIO'] / data['ASSETS']*100
data['deposit_share'] = data['DEPOSITS'] / data['LIABILITIES']*100
data['dep_to_assets'] = data['DEPOSITS'] / data['ASSETS']*100

data['NPLR'] = data['NPL'] / data['LOAN_PORTFOLIO']*100

data['CAR'] = data['EQUITY'] / data['ASSETS']*100
```

Также были рассчитаны такие показатели как:

9. Bank_Size – размер банка рассчитывается как натуральный логарифм от общих активов банка. Сглаживает разницу между очень крупными и мелкими банками и контролирует размер банка в одной модели.

10. Leverage – финансовый рычаг или мультипликатор собственного капитала, который рассчитывается как отношение активов к собственным средствам банка. Показывает степень использования заемных средств для финансирования активов. Чем выше финансовый рычаг, тем больше банк зависит от заемных средств. Если Leverage > 15% , то у банка высокая зависимость от заемных средств (выше доходность, но и выше риск дефолта), а если Leverage < 10%, то у банка консервативная политика (ниже доходность, но выше устойчивость). Также важно отметить, что высокий левередж усиливает чувствительность к прочим рискам.

11. CIR – отношение операционных расходов к операционным доходам. Если CIR меньше 1, то банк получает прибыль от своей операционной деятельности. Чем ниже CIR, тем меньше средств тратит банк на содержание персонала и административные расходы.

И прочие финансовые показатели, которые были опробованы в качестве контрольных переменных

```python
data['Bank_Size'] = np.log(data['ASSETS'])
data['Revenue'] = data['Interest_Income'] + data['Commission_Income'] + data['Operating_Income'] + data['Other_Income']
data['Leverage'] = data['ASSETS'] / data['EQUITY']
data['CIR'] = data['Operating_Expense'] / data['Operating_Income'] 
data['Net_Profit_Margin'] = data['Net_Profit'] / data['Revenue']*100
data['Asset_Turnover'] = data['Revenue'] / data['ASSETS']*100
data['Tax_Burder'] = data['Net_Profit'] / data['EBT']*100

# было 36 наблюдений, где DEPOSITS > LIABILITIES (видимо при расчтете обязательств), в этих наблюдениях deposit_share = dep_to_assets
mask = data['deposit_share'] > 1
data.loc[mask, 'deposit_share'] = data.loc[mask, 'DEPOSITS'] / (data.loc[mask, 'LIABILITIES'] + data.loc[mask, 'EQUITY'])

print(data.shape)
data.sample(5)
```

## Z-score

Также рассчитаем меру общего риска:

Z-score – показатель вероятности банкротства для банков. Рассчитывается как отношение суммы рентабельности активов (ROA) и отношения собственных средств к активам (CAR) к волатильности рентабельности активов (см. формула 1). Под волатильностью понимается стандартное отклонение от рентабельности активов. Показатель Z-score оценивает, насколько банк далек от дефолта. Чем выше Z-score, тем стабильнее банк.

```python
# Замена бесконечностей на NaN
data = data.replace([np.inf, -np.inf], np.nan)
```

```python
data['Z-score'] = (data['ROA'] + data['CAR']) / np.std(data['ROA'])
data['Z-score_RWA'] = (data['ROA'] + data['Н1.0']) / np.std(data['ROA'])
```

## Авторегрессия + средний прирост

Заполним пропуски последним имеющимся значением переменной + средний прирост за весь известный период по этой переменной для этого банка

```python
data = data.sort_values(by=['REGN', 'DATE'])

for var in var_with_null:
    # Рассчитываем прирост (разницу между текущим и предыдущим значением)
    data[f'{var}_diff'] = data.groupby('REGN')[var].diff()
    
    # Вычисляем средний прирост для каждого банка
    mean_diff = data.groupby('REGN')[f'{var}_diff'].mean().reset_index()
    mean_diff.rename(columns={f'{var}_diff': f'{var}_mean_diff'}, inplace=True)
    
    # Объединяем средний прирост с исходным DataFrame
    data = pd.merge(data, mean_diff, on='REGN', how='left')
    
    # Заполняем пропуски: текущее значение = предыдущее + средний прирост
    data[f'{var}_filled'] = data.groupby('REGN')[var].ffill() + data[f'{var}_mean_diff']

    data[var] = data[var].combine_first(data[f'{var}_filled'])
    data.drop([f'{var}_diff', f'{var}_mean_diff', f'{var}_filled'], axis=1, inplace=True)
    
data.sample(5)
```

## Первое будущее значение

Остались пропуски в тех наблюдениях, когда пропуски по показателю начинаются с 2011Q1 (начало датасета) по интересующий момент времени. Для таких наблюдений заполним пропуски первым "будущим" значением для данного банка по рассматриваемому показателю.

```python
def fill_with_first_future(group):
    # Находим индексы пропущенных значений
    null_mask = group.isnull()
    
    # Для каждого пропущенного значения ищем первое не-NaN значение после него
    for var in var_with_null:
        last_valid = group[var].last_valid_index()
        if last_valid is not None:
            group[var] = group[var].fillna(method='bfill')
    
    return group
data = data.groupby('REGN').apply(fill_with_first_future).reset_index(drop=True)
```

```python
## Заменим нули на 0.01, т.к. при сглаживании сезонности при делении на 0 образуются NaN
miss_0 = ((data_2022_im == 0).sum() / len(data_2022_im))*100
miss_0 = pd.DataFrame(round(miss_0, 2), columns = ['Share_of_missing, %'])
miss_0['Missing_count'] = (data_2022_im == 0).sum()
miss_0 = miss_0.sort_values(by='Share_of_missing, %', ascending=False)
miss_0 = miss_0.drop(['sys_importance', 'SANCTIONS', 'SANCTIONS_2014', 'SANCTIONS_2022',
        'SDN-list_2014', 'SSI-list_2014', 'SWIFT', 'SDN-list_2022'])
miss_0[miss_0.Missing_count > 0]
```

```python
cols = ['SDN-list_2014', 'SSI-list_2014', 'SWIFT', 'SDN-list_2022', 'SANCTIONS_2014', 'SANCTIONS_2022', 'SANCTIONS']

cols_to_replace = [col for col in data_2022_im.columns if col not in cols]
data_2022_im[cols_to_replace] = data_2022_im[cols_to_replace].replace(0, 0.01)
```

```python
numeric_cols = data_2022_im.select_dtypes(include=['number']).columns.difference(exclude_cols)

data_2022_im = data_2022_im.sort_values(['REGN', 'year', 'quarter'])

for col in numeric_cols:
    # Создаем временный мультииндекс для группировки
    temp_df = data_2022_im.set_index(['REGN', 'quarter', 'year'])[[col]].unstack()
    
    # Вычисляем YoY: текущий год / предыдущий год (только для пар "квартал + (year-1)")
    yoy = temp_df[col].div(temp_df[col].shift(axis=1))
    
    # Возвращаем в исходный формат
    yoy = yoy.stack().reset_index(name=f'{col}_YoY')
    
    # Мерджим обратно в основной датафрейм
    data_2022_im = data_2022_im.merge(
        yoy[['REGN', 'year', 'quarter', f'{col}_YoY']],
        on=['REGN', 'year', 'quarter'],
        how='left')
```
